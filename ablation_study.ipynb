```python
import numpy as np

import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn

from dataset import get_full_list, ChineseCharacterDataset

from models import CVAE, Attn_CVAE, Conv_CVAE, Res_CVAE, CAGAN

from utils import plot_generated_images, pad_to_target_size
```

```python
%load_ext autoreload
%autoreload 2
```

### Loading and Preprocessing of dataset:

```python
# Set the directory and load the dataset
image_dir = './chinese_chars/pngs'
full_data_list = get_full_list(image_dir)
```

```python
train_size = int(len(full_data_list))
print("Training Set Size:", train_size)
```

```python
train_data = ChineseCharacterDataset(full_data_list[:train_size], cond_type='Row', rows=[20,43])

# Create data loaders
train_loader = DataLoader(train_data, batch_size=32, shuffle=False)
```

### Loading model:

```python
# Use GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')
```

```python
input_dim = 64 * 64
z_dim = 50
condition_dim = 50
hidden_dim = 6400*8
learning_rate = 1e-3


model_CVAE = CVAE(input_dim, hidden_dim, z_dim, condition_dim).to(device)
model_Res_CVAE = Res_CVAE(z_dim, condition_dim).to(device)
model_Attn_CVAE = Attn_CVAE(input_dim, hidden_dim, z_dim, condition_dim).to(device)
model_Conv_CVAE = Conv_CVAE(z_dim, condition_dim).to(device)
model_CAGAN = CAGAN().to(device)
```

```python
# Create an iterator from the DataLoader
train_iter = iter(train_loader)

# Get one batch of data
one_batch = next(train_iter)

inputs, conditions = one_batch
conditions = conditions.to(device)
padded_condition = pad_to_target_size(conditions, (64, 64)).to(device)
```

```python
model_CVAE.initialize_fc_condition(padded_condition)
model_Attn_CVAE.initialize_fc_condition(padded_condition)
```

```python
model_CVAE.load_state_dict(torch.load('./saves/cvae_r20_43.pth')) 
model_Res_CVAE.load_state_dict(torch.load('./saves/resnet_cvae_r20_43.pth')) 
model_Attn_CVAE.load_state_dict(torch.load('./saves/attn_cvae_r20_43.pth')) 
model_Conv_CVAE.load_state_dict(torch.load('./saves/unet_cvae_r20_43.pth')) 
model_CAGAN.load_state_dict(torch.load('./saves/ca_gan_r20_43.pth')) 
```

### Generating Images:

```python
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
```

```python
import matplotlib.pyplot as plt

model_CVAE.eval()
model_Res_CVAE.eval()
model_Attn_CVAE.eval()
model_Conv_CVAE.eval()
model_CAGAN.eval()

batch_size = 32
z_dim = 50
print_images = 5

with torch.no_grad():
    for batch_idx, (data, condition) in enumerate(train_loader):
        data = data.to(device)
        condition = condition.to(device)
        padded_condition = pad_to_target_size(condition, (64, 64)).to(device)
        
        z = torch.randn(batch_size, z_dim).to(device)
        
        CVAE_cond = model_CVAE.forward_condition_encoder(padded_condition)
        CVAE_sample = model_CVAE.decode(z, CVAE_cond).cpu()
        CVAE_sample = CVAE_sample.view(batch_size, 1, 64, 64)
        
        Res_CVAE_sample = model_Res_CVAE.decoder(z, condition).cpu()
        Res_CVAE_sample = Res_CVAE_sample.view(batch_size, 1, 64, 64)
        
        Attn_CVAE_cond = model_Attn_CVAE.forward_condition_encoder(padded_condition)
        Attn_CVAE_sample = model_Attn_CVAE.decode(z, Attn_CVAE_cond).cpu()
        Attn_CVAE_sample = Attn_CVAE_sample.view(batch_size, 1, 64, 64)
        
        Conv_CVAE_sample = model_Conv_CVAE .decode(z, padded_condition).cpu()
        Conv_CVAE_sample = Conv_CVAE_sample.view(batch_size, 1, 64, 64)

        z_gan = torch.randn(batch_size, 1024).to(device)
        CAGAN_cond = torch.cat([z_gan, padded_condition.flatten(1)], dim=1)
        CAGAN_sample = model_CAGAN.decoder(CAGAN_cond).cpu()
        CAGAN_sample = CAGAN_sample.view(batch_size, 1, 64, 64)

        for i in range(print_images):
            ref = data[i].cpu().detach().numpy().reshape(64, 64)
            cond = condition[i].cpu().detach().numpy()
            if len(cond.shape) > 2:
                cond = cond[0]  # Select the first channel if condition is multi-channel
            cond_resized = cond.reshape(cond.shape)

            CVAE_img = CVAE_sample[i].cpu().detach().numpy().reshape(64, 64)
            Res_CVAE_img = Res_CVAE_sample[i].cpu().detach().numpy().reshape(64, 64)
            Attn_CVAE_img = Attn_CVAE_sample[i].cpu().detach().numpy().reshape(64, 64)
            Conv_CVAE_img = Conv_CVAE_sample[i].cpu().detach().numpy().reshape(64, 64)
            CAGAN_img = CAGAN_sample[i].cpu().detach().numpy().reshape(64, 64)
                
            plt.figure(figsize=(18, 4))

            # Plot condition image
            plt.subplot(1, 7, 1)
            plt.title('Condition Image')
            plt.imshow(cond_resized, cmap='gray')
            plt.axis('off')

            # Plot reference image
            plt.subplot(1, 7, 2)
            plt.title('Reference Image')
            plt.imshow(ref, cmap='gray')
            plt.axis('off')

            # Plot CVAE generated image
            plt.subplot(1, 7, 3)
            plt.title('CVAE')
            plt.imshow(CVAE_img, cmap='gray')
            plt.axis('off')

            # Plot Res_CVAE generated image
            plt.subplot(1, 7, 4)
            plt.title('Res_CVAE')
            plt.imshow(Res_CVAE_img, cmap='gray')
            plt.axis('off')

            # Plot Attn_CVAE generated image
            plt.subplot(1, 7, 5)
            plt.title('Attn_CVAE')
            plt.imshow(Attn_CVAE_img, cmap='gray')
            plt.axis('off')

            # Plot Conv_CVAE generated image
            plt.subplot(1, 7, 6)
            plt.title('Conv_CVAE')
            plt.imshow(Conv_CVAE_img, cmap='gray')
            plt.axis('off')

            # Plot CAGAN generated image
            plt.subplot(1, 7, 7)
            plt.title('CAGAN')
            plt.imshow(CAGAN_img, cmap='gray')
            plt.axis('off')

            plt.show()
        break
```
